## Teorema di Shannon

Dato un segnale $s\left(t\right)$ la cui banda è limitata dalla frequenaza $f_M$, allora $s\left(t\right)$ può essere univocamente ricostruito dai suoi campioni presi a una frequenza che rispetta la condizione $f_c>2f_M$

#### Simulazione

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import ipywidgets as wdg

TIME_W = 1e-3  #larghezza della finestra
Fc = 5e5       #frequenza di ccampionamento
Tc = 1/Fc      #tempo di campionamento
Nc = Fc*TIME_W #Numero di campioni

ax_t = np.linspace(0,TIME_W, Fc) #asse dei tempi Fc punti deve dare l'effetto di una sinusoide analogica
ax_tc = np.linspace(0,TIME_W,Nc) #asse dei tempi campionato

f1 = 2e5
x1 = np.sin(2*np.pi*f1*ax_t)
x1c = np.sin(2*np.pi*f1*ax_tc)

f2 = 3e5
x2 = np.sin(2*np.pi*f2*ax_t)
x2c = np.sin(2*np.pi*f2*ax_tc)
```

```python
gs = gridspec.GridSpec(3, 2)

figure = plt.figure(figsize=(20,20))
plt.subplot(gs[0,0])
plt.plot(ax_t,x1)
plt.title("$sin(2\pi{200*10^3}t)$",fontsize=22, fontweight='bold')

plt.subplot(gs[0,1])
plt.plot(ax_t,x2)
plt.title("$sin(2\pi{300*10^3}t)$",fontsize=22, fontweight='bold')


plt.subplot(gs[1,0])
plt.stem(ax_tc,x1c)
plt.title("$sin(2\pi{200*10^3}T_c)$",fontsize=22, fontweight='bold')

plt.subplot(gs[1,1])
plt.stem(ax_tc,x2c)
plt.title("$sin(2\pi{300*10^3}T_c)$",fontsize=22, fontweight='bold')

plt.subplot(gs[2,0])
X1c = np.fft.fft(x1c)
axf = np.linspace(0.0, 1/(2*Tc), int((TIME_W*Fc))//2)
plt.plot(axf,np.abs(X1c[0:(int((TIME_W*Fc))//2)]))
plt.title("$FFT\{sin(2\pi{200*10^3}T_c)\}$",fontsize=22, fontweight='bold')

plt.subplot(gs[2,1])
X2c = np.fft.fft(x2c)
axf = np.linspace(0.0, 1.0/(2*Tc), 500//2)
plt.plot(axf,np.abs(X2c[0:(500//2)]))
plt.title("$FFT\{sin(2\pi{300*10^3}T_c)\}$",fontsize=22, fontweight='bold')

plt.tight_layout()
figure.subplots_adjust(top=0.85)
plt.show()
```

```python
def verify_shannon(f3):
    x3 = np.sin(2*np.pi*f3*ax_t)
    x3c = np.sin(2*np.pi*f3*ax_tc)
    
    gs = gridspec.GridSpec(1, 2)

    figure = plt.figure(figsize=(25,5))

    plt.subplot(gs[0,0])
    X3c = np.fft.fft(x3c)
    axf = np.linspace(0.0, 1.0/(2*Tc), 500//2)
    plt.plot(axf,np.abs(X3c[0:(500//2)]))
    plt.title("$FFT\{sin(2\pi{f_{3}}T_c)\}$",fontsize=22, fontweight='bold')

    plt.tight_layout()
    figure.subplots_adjust(top=0.85)
    plt.show()
    
wdg.interact(verify_shannon, f3=(0,5e5,1e4))
```

#### Svolgimento
Abbiamo generato la sinusoide $x\left(t\right)=sin(2\pi f_0 t)$ con $f_0 = 200KHz$ e ne abbiamo visualizzato lo spettro.

Il modulo dello spettro della sinusode risulta essere una delta di Dirac definita in $f_0$, variandone la frequenza la delta si centra nella nuova pulsazione.

Quando la frequenza della sinusoide supera i 250kHz lo spettro visualizzato non coincide piu con lo spettro del segnale originale, da cui anche l'impossibilità di ricostruirlo. Ciò accade perchè l'oscilloscopio effettua un campionamento sul segnale a 500kHz. 

Se la frequenza fondamentale del segnale di ingresso supera i 250KHz non è rispettato il teorema di Shannon e sull'oscilloscopio si visualizzano gli alias delle frequenze del segnale cioè le repliche periodiche dello spettro del segnale d'ingresso centrate alla frequenza di campionamento che risultano essre frequenze non presenti nel segnale originale.

Una sinusoide a 200kHz risulta quindi indistinguibile da una sinusoide a 300kHz se campionate a 500kHz.

## Influenza della finestra di acquisizione

Per valutare l'influenza della finestra temporale sul segnale abbiamo utilizzato due segnali periodici:
- un onda quadra
- un segnale sinusoidale

### Onda Quadra

```python
t = np.linspace(-2,2,801)
r = 1*(abs(t)<1/2)+1/2*(abs(t)==1/2)
p = 1*((t%2)==0)
fig = plt.figure(figsize=(20,10))
plt.plot(t,np.convolve(r,p,'same'))
plt.xticks([-2,-1.5,-1,-0.5,0,0.5,1,1.5,2],["-2T","-(3/2)T","-T","-T/2","0","T/2","T","(3/2)T","2T"],fontweight='bold')
plt.show()
```

Un onda quadra è un segnale periodico che può essere ottenuto a partire da un segnale rettangolare di durata T/2 e un pettine di delta  di periodo T attraverso un operazione di convoluzione. In frequenza si ottiene il prodotto delle trasformate di Fourier quindi il prodotto tra un pettine di delta di periodo 1/T e riscalato ed un seno cardinale che attraversa lo zero nei multipli di 2/T, quindi nella trasformata di Fourier dell'onda saranno presenti solo i multipli dispari della frequenza fondamentale.

Inoltre poiche il segnale presenta discontinuità la sua trasformata decresce come 1/f 

```python
r = np.sinc(t)
p = 1*((t%0.5)==0)
fig = plt.figure(figsize=(20,10))
plt.plot(t,r)
plt.plot(t,p)
plt.xticks([-2,-1.5,-1,-0.5,0,0.5,1,1.5,2],["-4T","-3/T","-2/T","-1/T","0","1/T","2/T","3/T","4/T"],fontweight='bold')
plt.show()
```

Gli spettri visualizzati sull'oscilloscopio non coincidono con quelli calcolati teoricamente, questo avviene principalmente poichè il segnale sull'oscilloscopio è visualizzato in un determinato intervallo di tempo, quindi non stiamo valutando il segnale originale ma una sua versione finestrata. Esistono vari tipi di finestre con varie caratteristiche. Noi abbiamo utilizzato le seguenti finestre:

Osseriviamo come la finestra rettangolare sia la più selettiva in frequenza ed è anche quella che altera meno l'ampiezza della prima armonicha. La selettività in frequenza è data dalla larghezza del lobo principale del seno cardinale, se il rettangolo di acquisizione è largo N allora il suo spettro intercetta lo zero nei multipli di N quindi il lobo principale è largo 2N. Questa caratteristica determina la selettività in frequenza cioè quanto è possibile distinguere due frequenze vicine. 

In ordine le finestre con maggiore selettività sono rettangolare, Hanning, BlackmanHarris è FlatTop.

Uno svantaggio della finestra rettangolare è che presenta discontinuità è quindi nello spettro si introducono oscillazioni che hanno un entità che dipende dalla larghezza della finestra, più è larga più queste oscillazioni decadono velocemente ma sono di entità maggiore.

La finestra rettangolare e quella di Hanning inducono oscillazioni molto marcate, la Flat top e quella con oscillazioni meno marcate e una via di mezzo è la BlackmanHarris.


# Segnale Gradino

Ci siamo proposti di studiare lo spettro di un segnale gradino che nei tempi presenta una discontinuità. Con il generatore di funzioni non è possibile creare tale segnale quindi abbiamo generato un onda quadra e l'abbiamo visualizzata sull'oscilloscopio restringendo molto l'intervallo temporale di visualizzazione in queto modo sull'oscilloscopio si visualizza un gradino.

Nei tempi un gradino è descritto dalla seguente equazione $\varepsilon\left(t\right) = \left\{\begin{matrix}0&t<0\\\frac{1}{2}&t=0\\1&t>0\end{matrix} \right.$

Applicando la definizione si puo calcolare la sua trasformata di Fourier: $\varepsilon\left(f\right) = \int\limits^\infty_{-\infty}\varepsilon\left(f\right)e^{-j\pi ft}dt = 
\int\limits^\infty_{0}e^{-j\pi ft}dt = \frac{1}{-J2\pi f}\left.e^{-j\pi ft}\right|_{0}^{\infty} = \frac{1}{j2\pi f} $

In frequenza risulta essere un'iperbole equilatera.

## Modulazione AM

Nella modulazione AM si utilizza un segnale in bassa frequenza detto modulante per variare l'ampiezza di un segnale in alta frequenza detto portante.

$c\left(t\right) = Acos\left(2\pi f_c t\right)$

$m\left(t\right) = Bcos\left(2\pi f_m t\right)$

Il segnale modulato risulta avere la seguente espressione analitica $s_{AM}\left(t\right) = A\left[1 + m_acos\left(2\pi f_m t\right)\right]cos\left(2\pi f_c t\right)$

## Modulazione FM

Nella modulazione FM si utilizza un segnale in bassa frequenza detto modulante per variare la frequenza istantanea di un segnale detto portante.

La frequenza istantanea del segnale modulato vale $f_i\left(t\right) = f_c + k_fm\left(t\right)$  
dove

- $k_{f}$ è una costante che viene detta fattore di sensibilità in frequenza del modulatore
- $m\left(t\right)$ è il segnale modulante
- $f_{c}$ è la frequenza della portante non modulata

$\theta_i\left(t\right) = 2\pi\int_{0}^{t}f_i\left(\tau\right)d\tau = 2\pi f_ct + 2\pi k_f\int_{0}^{t}m\left(\tau\right)d\tau $

nel caso in cui il segnale modulate sia una cosinusoide allora il segnale modulato risulta

$s_{FM}\left(t\right) = A_c cos\left( 2\pi f_ct + \frac{A_mk_f}{f_m} sin(2\pi f_m t) \right) = A_c cos\left( 2\pi f_ct + \frac{\Delta f}{f_m} sin(2\pi f_m t) \right) $

#### Regola di Carson
Un segnale modulato in frequenza teoricamente occupa una banda illimitata ma si può osservare che a una certa distanza dalla portante i contriuti spettrali diventano via via piu trascurabili. La banda di un segnale FM puuò quindi essere ben approssimata attraverso la regla di Carson: $B_{FM} = 2\Delta{f}+2f_m$

Abbiamo generato un segnale modulato FM con il generatore di funzioni con frequenzas portante pari a 1MHz e frequenza modulante pari a 10KHz e deviazione di frequenza di 70KHz. Ne abiamo visualizzato lo spettro e valutato la validità della formula di Carson

## Conclusioni

Il DSP puo portare a effetti indesiderati tipo un campionamento a una frequenza inadeguata può aggiungere/togliere componenti frequenziali dal segnale e renderlo completamente diverso dal segnale originale.

### Finetra
Quando si effettua l'acquisizione di un segnale va scelta la finestra adeguata in base alle caratteristiche del segnale.

### AM
La modulazione AM ha una buona occupazione di banda che puo essere molto migliorata, infatti il segnale di informazione è presente in due bande frequenziali simmetriche rispetto la frequenza portante, in trasmissione può essere trasmessa una sola di queste tenendo conto che dovrà essere corretta l'ampiezza del segnale demodulato in questo modo dimezza l'occupazione di banda e la potenza da trasmettere.

Secondo lo stesso principio poichè la portante non contiene informazione si può decidere di non trasmetterla risparmiando ulteriore energia in trasmissione.

Questa modulazione lavorando sull'ampiezza dei segnali risulta essere molto debole rispetto al rumore.

### FM
La modulazione FM occupa molta più banda rispetto a quella AM, e può essere calcolata approssimata molto bene con la formula di Carson.

Questa modulazione operando sulla frequenza del segnale risulta essere molto più robusta rispetto al rumore.
