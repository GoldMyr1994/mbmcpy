```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def inv_color(img):
    nimg = np.zeros_like(img)
    nimg[:,:,0] = img[:,:,2]
    nimg[:,:,1] = img[:,:,1]
    nimg[:,:,2] = img[:,:,0]
    return nimg
```

```python
img = cv2.imread('1.png')[:,1510:3400,:]
plt.figure(figsize=(20,10))
plt.imshow(inv_color(img))
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
list_c = []

for e in img:
    for pix in e:
        if pix in list_c: continue
        else:
            print pix
            list_c.append(pix)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
plt.imshow(inv_color(img)[:,:,0],'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2)
plt.imshow(inv_color(img)[:,:,1],'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3)
plt.imshow(inv_color(img)[:,:,2],'gray')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite("rosso.png", inv_color(img)[:,:,0] );
cv2.imwrite("verde.png", inv_color(img)[:,:,1] );
cv2.imwrite("blu.png", inv_color(img)[:,:,2] );
```

```python
image = np.zeros_like(img)
for i in range(img.shape[0]/2):
    for j in range(img.shape[1]/2):
        if(img[i,j,0]<100 and img[i,j,1]<70 and img[i,j,2]<100): 
            image[i,j,:] = 0
            continue
        if(img[i,j,0]==92 and img[i,j,1]==92 and img[i,j,2]==92):
            image[i,j,:] = 0
            continue
        if(img[i,j,0]==122 and img[i,j,1]==122 and img[i,j,2]==122):
            image[i,j,:] = 0
        else: image[i,j,:] = 255
```

```python
plt.figure(figsize=(20,10))
plt.imshow(inv_color(image))
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite("prova1.png", image );
```
