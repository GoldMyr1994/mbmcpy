```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TIME_W = 1e-3  #larghezza della finestra
Fc = 5e5       #frequenza di ccampionamento
Tc = 1/Fc      #tempo di campionamento
Nc = Fc*TIME_W #Numero di campioni

ax_t = np.linspace(0,TIME_W, Fc) #asse dei tempi Fc punti deve dare l'effetto di una sinusoide analogica
ax_tc = np.linspace(0,TIME_W,Nc) #asse dei tempi campionato

f1 = 2e5
x1 = np.sin(2*np.pi*f1*ax_t)
x1c = np.sin(2*np.pi*f1*ax_tc)

f2 = 3e5
x2 = np.sin(2*np.pi*f2*ax_t)
x2c = np.sin(2*np.pi*f2*ax_tc)
```

```python
fig_1 = plt.figure(figsize=(40,40))
plt.subplot(4,2,1)
plt.title("$sin(2\pi{200*10^3}t)$");
plt.plot(ax_t,x1)
plt.subplot(4,2,2)
plt.title("$sin(2\pi{200*10^3}T_c)$");
plt.stem(ax_tc,x1c)
plt.subplot(4,2,3)
plt.title("$sin(2\pi{300*10^3}t)$");
plt.plot(ax_t,x2)
plt.subplot(4,2,4)
plt.title("$sin(2\pi{300*10^3}T_c)$");
plt.stem(ax_tc,x2c)
plt.tight_layout()
plt.show()
```

```python
fig_1 = plt.figure(figsize=(18,9))
plt.subplot(1,2,1)
X1c = np.fft.fft(x1c)
axf = np.linspace(0.0, 1/(2*Tc), int((TIME_W*Fc))//2)
plt.title("$FFT\{sin(2\pi{200*10^3}T_c)\}$");
plt.plot(axf,np.abs(X1c[0:(int((TIME_W*Fc))//2)]))
plt.subplot(1,2,2)
X2c = np.fft.fft(x2c)
axf = np.linspace(0.0, 1.0/(2*Tc), 500//2)
plt.title("$FFT\{sin(2\pi{300*10^3}T_c)\}$");
plt.plot(axf,np.abs(X2c[0:(500//2)]))
plt.show()
```

Il campionamento su un segnale introduce repliche periodiche dello spettro base del segnale attorno ai multipli interi della frequenza di campionamento.


<p>Il risultato è una replica dello spettro del seganele originale intorno a tutti i multipli della frequenza di
campionamento<p>

<p>se il segnale è una sinusoide il modulo della trasformata presenta due delta una in $F_0$ e l'altra in $-F_0$</p>
<p>la convoluzine con un pettine di delta di periodo $F_C$ ritorna una delta in $F_C-F_0$ e una in $F_C+F_0$</p>
<p>Con due sinusoidi, una alla frequenza di 200KHz e una a 300KHz campionate alla frequnza di 500KHz si possono osservare i risultati spettrali(sopra grafici) </p>
<p>i due spettri sono uguali ma attenzione il primo grafico rappresen la componente spettrale della sinusoide a 200KHz, la seconda immagine ..... è la replica spettrale negativa diciamo 500000-300000</p>
